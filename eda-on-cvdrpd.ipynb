# Forming the data into mathimatical form

```python
# checking the categorical features
cols = ['General_Health','Exercise','Heart_Disease','Skin_Cancer','Other_Cancer','Depression','Diabetes','Arthritis','Sex','Age_Category','Smoking_History']
for i in cols:
    print(data[i].value_counts())
    print()
```

**First replace Yes and No with 1,0 respectively other than these columns 'Diabetes','Sex','Age_Category' and 'General_Health' because they have different categorical values**

```python
cols_1 = ['Exercise','Heart_Disease','Skin_Cancer','Other_Cancer','Depression','Arthritis','Smoking_History']
data[cols_1] = data[cols_1].replace(['Yes','No'],[1,0])

data[cols_1].head()
```

# 1. General Health

```python
# Now General Health
data['General_Health'].value_counts()
```

```python
data['General_Health'] = data['General_Health'].replace(['Very Good','Good','Excellent','Fair','Poor'],[5,4,3,2,1])
data['General_Health']
```

# 2. Sex

```python
# Now Sex
data['Sex'].value_counts()
```

```python
data['Sex'] = data['Sex'].replace(['Male','Female'],[1,0])
data['Sex']
```

# 3. Checkup

```python
# Now Checkup
data['Checkup'].value_counts()
```

```python
data['Checkup'] = data['Checkup'].replace(
    [
        'Within the past year',
        'Within the past 2 years',
        'Within the past 5 years',
        '5 or more years ago',
        'Never'
    ],[1,2,5,6,0])
data['Checkup']
```

# 4. Age Category

```python
#Now Age Category
data['Age_Category'].value_counts()
```

```python
data['Age_Category'] = data['Age_Category'].str[:2].astype(int)
data['Age_Category']
```

# 5. Diabetes
0 : No
1 : Yes
10 : No, Borderline
11 : Yes, during Pregnancy

```python
#now Diabetes
data['Diabetes'].value_counts()
```

```python
data['Diabetes'] = data['Diabetes'].replace(
    [
        'No',
        'Yes',
        'No, pre-diabetes or borderline diabetes',
        'Yes, but female told only during pregnancy'
    ],[0,1,10,11])
data['Diabetes']
```

```python
heart_patient_plus = heart_patient[(heart_patient.Skin_Cancer == 1) | (heart_patient.Other_Cancer == 1) | (heart_patient.Depression == 1) | (heart_patient.Diabetes == 1) | (heart_patient.Arthritis == 1)]
heart_patient_plus
```

```python
x = { 'Total Heart Patient' : len(heart_patient), 'Only Heart Patient' : len(heart_patient) - len(heart_patient_plus),'Heart Patient Plus' : len(heart_patient_plus)}
patient_plus = pd.DataFrame(x, index = [0])
sns.barplot(patient_plus)
```

**Here we can crealy see that Most of the Heart Patient having other disease also.
It means these diseases are triggring each other**

```python
normal_weight = len(heart_patient[(heart_patient.BMI <= 25.0) & (heart_patient.BMI > 18.0)])
under_weight = len(heart_patient[(heart_patient.BMI <= 18)])
over_weight = len(heart_patient[(heart_patient.BMI >= 25)])
x = { 'Total' : len(heart_patient),'Normal Weight' : normal_weight, 'Under Weight' : under_weight,'Over Weight' : over_weight }
bmi_based = pd.DataFrame(x, index = [0])
sns.barplot(bmi_based)
```
